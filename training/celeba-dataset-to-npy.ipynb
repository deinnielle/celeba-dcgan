```
import numpy as np
import zipfile
from PIL import Image
import imageio
import os
```

```
from google.colab import drive
drive.mount('./mount')
```

```
with zipfile.ZipFile('mount/My Drive/Colab Notebooks/celeba-dcgan/celeba-dataset.zip', 'r') as zf:
  images_to_extract = 20000
  image_array = []
  files_in_zip = zf.namelist()

  for i, image in enumerate(files_in_zip[1:images_to_extract + 1]):
    image_file = zf.extract(image)
    image = Image.open(image_file)
    crop_image = image.crop((0, 20, 178, 218 - 20))
    resize_image = crop_image.resize((128, 128))
    resize_image.save('cropped-and-resized.png')
    read_image = imageio.imread('cropped-and-resized.png')
    image_array.append(read_image)
    os.remove(image_file)

    if (i%1000 == 0):
        print(i, ' done..')
        
  os.remove('cropped-and-resized.png')
```

```
np.save('mount/My Drive/Colab Notebooks/celeba-dcgan/celeba-' + str(images_to_extract) + '.npy', image_array)
```
